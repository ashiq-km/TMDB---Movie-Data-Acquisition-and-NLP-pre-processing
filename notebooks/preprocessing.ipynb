Next we will remove the stop words from this Description.



```python
from nltk.corpus import stopwords
```

```python
sw_eng = stopwords.words('english')
```

```python
def r_sw(text):
    result = []
    words = text.split()
    for i in words:
        if i in sw_eng:
            result.append("")
        else:
            result.append(i)
    return " ".join(result)
```

```python
data['Description'] = data['Description'].apply(r_sw)
```
